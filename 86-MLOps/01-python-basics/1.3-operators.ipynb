#### Logical Operators

```python
## And, Not, OR
X=True
Y=True

result = X and Y
print(result)
```

```python
## And, Not, OR
X=False
Y=True

result = X and Y
print(result)
```

```python
## OR
X=False
Y=False

result = X or Y
print(result)
```

```python
# Not operator
X=False
not X
```
