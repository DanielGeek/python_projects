### Variables
Variables are fundamental elements in programming used to store data that can be referenced and manipulated in a program. In Python, variables are created when you assign a value to them, and they do not need explicit declaration to reserve memory space. The declaration happens automatically when you assign a value to a variable.

```python
## Declaring And Assigning Variables

age=35
height=1.86
name="Daniel"
is_student=True

## printing the variables

print("Age:", age)
print("Height:", height)
print("Is Student:", is_student)
print("Name:", name)
```

```python
## Naming Conventions
## Variable names should be descriptive
## They must start with a letter or an '_' and contains letter, number, and underscores
## Variables names case sensitive

# valid variable names
first_name = "Daniel"
last_name = "Ángel"
```

```python
# Invalid variable names
#2age=30
# first-name="Daniel"
#@name="Daniel"
```

```python
## Case sensitivity
name="Daniel"
Name="Elías"

name==Name
```

```python
## Undestanding Variable Types
## Python is dynamically, typed, type of a variable is determined at runtime
age=35 # Integer
height=1.86 # Float
name="Daniel" # String
is_student=True # Boolean

print(type(name))
```

```python
## Type Checking and Conversion

type(height)
```

```python
age=25
print(type(age))

# Type conversion
age_str=str(age)
print(age_str)
print(type(age_str))
```

```python
## Dynamic Typing
## Python allows you to change the type of a variable by reassigning it to a different value

var=10 #int
print(var, type(var))

var="Hello"
print(var, type(var))

var=3.14
print(var, type(var))
```

```python
## Input

age=int(input("What is your age?"))
print("You entered:", age, type(age))
```

```python
### Simple calculator
num1=float(input("Enter first number: "))
num2=float(input("Enter second number: "))

sum = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2

print("Sum:", sum)
print("Difference:", difference)
print("Product:", product)
print("Quotient:", quotient)
```
