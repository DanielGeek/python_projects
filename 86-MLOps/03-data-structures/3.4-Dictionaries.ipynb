##### Introduction to Dictionaries

Dictionaries are unordered collections of items. They store data in key-value pairs.
Keys must be unique and immutable (e.g., strings, numbers, or tuples), while values can be of any type.

```python
## Creating Dictionaries
empty_dict={}
print(type(empty_dict))
```

```python
empty_dict=dict()
empty_dict
```

```python
student={"name": "Daniel", "age": 35, "grade": 24}
print(student)
print(type(student))
```

```python
# Single key is always used
student={"name": "Daniel", "age": 35, "name": 24}
print(student)
```

```python
## Accesing Dictionary Elements
student={"name": "Daniel", "age": 35, "grade": "A"}
print(student)
```

```python
## Accesing Dictionary elements
print(student["grade"])
print(student["age"])

## Accessing using get() method
print(student.get("grade"))
print(student.get("last_name"))
print(student.get("last_name", "Not Available"))
```

```python
## Modifying Dictionary Elements
## Dictionary are mutable, so you can add, update or delete elements
print(student)
```

```python
student["age"]=35 ## Update value for the key
print(student)
student["address"]="Venezuela" ## Added a new key and value
print(student)
```

```python
del student["grade"] ## Delete key and value pair
print(student)
```

```python
## Dictionary methods

keys = student.keys() ## get all the keys
print(keys)
values = student.values() ## get all values
print(values)

items = student.items() ## Get all key value pairs
print(items)
```

```python
## Shallow copy
student_copy = student
print(student)
print(student_copy)
```

```python
student["name"] = "Daniel2"
print(student)
print(student_copy)
```

```python
student_copy1 = student.copy() ## Shallow copy
print(student_copy1)
print(student)
```

```python
student["name"] = "Daniel3"
print(student_copy1)
print(student)
```

```python
### Interating Over Dictionaries
### You can use loops to iterate over dictionaries, keys, values, or items

## Iterating over keys
for keys in student.keys():
    print(keys)
```

```python
## Iterate over values
for values in student.values():
    print(values)
```

```python
## Iterate over key value pairs
for key, value in student.items():
    print(f"{key}: {value}")
```

```python
## Nested Dictionaries
students = {
    "student1" : { "name": "Daniel", "age": 35 },
    "student2" : { "name": "Luis", "age": 32 }
}

print(students)
```

```python
## Access nested dictorionaries elements
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
students.items()
```

```python
## Iterating over nested dictorionaries
for student_id, student_info in students.items():
    print(f"{student_id}: {student_info}")
    for key, value in student_info.items():
        print(f"{key}: {value}")
```

```python
## Dictionary Comphrenhension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## Condition dictionary comprehension
events={x:x**2 for x in range(10) if x%2==0}
print(events)
```

```python
## Practical Examples

## Use a dictionary to count he frequency of elements in list

numbers=[1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
    print(frequency)
```

```python
## Merge 2 dictionaries into one

dict1={"a": 1, "b": 2}
dict2={"b": 3, "c": 4}
merged_dict={**dict1, **dict2}
print(merged_dict)
```
